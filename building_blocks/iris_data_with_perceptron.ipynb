```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
import matplotlib.pyplot as plt 
from perceptron import Perceptron
```

```python
iris_data=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",header=None,encoding="utf-8")
```

```python
iris_data.tail()
```

```python
X=iris_data.iloc[:,:-1].values
y=iris_data.iloc[:,-1].values
```

```python
y=np.where(y=="Iris-virginica",0,1)
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,3],c=y)
```

```python
print(y)
```

```python
iris_classifier=Perceptron(eta=0.01,n_iter=1000)
iris_classifier.fit(X,y)
```

```python
predicted_out=[]
for x in X:
    predicted_out.append(iris_classifier.predict(x))
```

```python
error=0
```

```python
for x,target in zip(X,y):
    predict_output=iris_classifier.predict(x)
    loss=target-predict_output
    error =error+ int(loss!=0.0)
```

```python
error
```

```python
predicted_out==y
```

```python
plt.plot(range(1,len(iris_classifier.errors)+1),iris_classifier.errors)
```
